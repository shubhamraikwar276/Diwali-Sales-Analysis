```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

ax = sns.countplot(x = 'Gender', data = df)
```

```python
#Count of Male vs Female

ax = sns.countplot(x = 'Gender', data = df, hue = 'Gender')
plt.title('Distribution of Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.columns
```

```python
sales_gen = df[['Gender','Amount']].groupby('Gender').sum().reset_index()

```

```python
sales_gen
```

```python
#Purchasing power
sns.barplot(x = 'Gender', y = 'Amount', data =sales_gen, hue = 'Gender')
plt.title("Purchasing Power of Different Gender")
```

From the above graphs we can see that most of the buyers are females and they also have higher purchasing power.

```python
#age group of male female count

bx = sns.countplot(x = "Age Group", data  = df, hue = 'Gender')

plt.title('Purchase in Age Group')

for bars in bx.containers:
    bx.bar_label(bars)
```

```python
#Total amount vs Age group

sales_amt = df[['Amount','Age Group']].groupby('Age Group').sum().sort_values(by = 'Amount', ascending = False).reset_index()
```

```python
sales_amt
```

```python
cx = sns.barplot(x = 'Age Group', y = 'Amount', data = sales_amt, hue = 'Age Group')
plt.title("Most Amt of purchase in Age Groups")
```

From this graph, we can see that females in all groups are more than males
But most importantly females between 26 - 35 age group are more likely to purchase among other age groups.

```python
#total number of orders from top 10 states

ord_state = df[['State','Orders']].groupby('State').sum().sort_values(by = 'Orders', ascending = False).head(10).reset_index()
```

```python
sns.barplot(x = 'State', y = 'Orders', hue = 'State', data = ord_state)
sns.set(rc = {'figure.figsize':(10,5)})
```

```python
#total amount of orders from top 10 states
amt_state = df[['State','Amount']].groupby('State').sum().sort_values(by = 'Amount', ascending = False).head(10).reset_index()

amt_state
```

```python
sns.barplot(x = 'State', y = 'Amount', data = amt_state, hue = 'State')
sns.set(rc = {'figure.figsize':(15,5)})
```

From the above graphs we can say that most of the orders are from - Uttar Pradesh, Maharashtra, Kerala

```python
#Marital Status count

cx = sns.countplot(x = 'Marital_Status', data = df, hue = 'Marital_Status')
sns.set(rc = {'figure.figsize': (7,5)})

for bars in cx.containers:
    cx.bar_label(bars)
```

```python
df.columns
```

```python
cx = sns.countplot(x = 'Marital_Status', data = df, hue = 'Gender')
sns.set(rc = {'figure.figsize': (7,5)})

for bars in cx.containers:
    cx.bar_label(bars)
```

```python
#Amount vs Marital Status
sns.barplot(x = 'Marital_Status', y = 'Amount', estimator = 'sum', data = df, hue = 'Gender')
plt.title("Amount vs Marital_Status")
```

From the above graphs we can see that, mostly married females are tend to purchase more and have more
purchasing power.

```python
#Occupation counts
sns.set(rc = {'figure.figsize': (20,5)})
dx = sns.countplot(x = 'Occupation', data = df, hue = 'Occupation')

for bars in dx.containers:
    dx.bar_label(bars)
```

```python
df.columns
```

```python
amt_occupation = df[['Amount','Occupation']].groupby('Occupation').sum().sort_values(by = 'Amount', ascending = False).reset_index()
```

```python
amt_occupation
```

```python
sns.barplot(x = 'Occupation', y = 'Amount', data = amt_occupation, hue = 'Occupation')
```

Most of the buyers are from IT Sector, Healthcare and Aviation and also most purchasing power also belongs to these category.

```python
#Product Category

ex = sns.countplot(data = df, x = 'Product_Category', hue = 'Product_Category')
sns.set(rc = {'figure.figsize': (25,5)})

for bars in ex.containers:
    ex.bar_label(bars)
```

```python
prd_amt = df[['Product_Category','Amount']].groupby("Product_Category").sum().sort_values(by = 'Amount', ascending = False).reset_index().head(10)
```

```python
sns.barplot(x = 'Product_Category', y = 'Amount', data = prd_amt, hue = 'Product_Category')
```

```python
prd_10 = df[['Product_ID','Orders']].groupby('Product_ID').sum().sort_values(by = 'Orders', ascending = False).head(10).reset_index()
```

```python
prd_10
```

```python
#Top ten products sold in numbers.
sns.barplot(x = 'Product_ID', y = 'Orders', data = prd_10, hue = 'Product_ID')
sns.set(rc = {'figure.figsize': (15,5)})
```

From the above graphs we can see that Food, Clothing & Appliances, Electronics & Gadgets are most sold.

CONCLUSION:

Married Females from Uttar Pradesh, Maharashtra, Kerala working in sectors - IT Sector, Healthcare and Aviation are more likely to purchase products of the category - Food, Clothing & Appliances, Electronics & Gadgets.
